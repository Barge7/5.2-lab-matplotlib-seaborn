```python
# sns.histplot(x= titanic.Age, element= 'step')
plt.axvline(titanic.Age.mean(), c= 'red', linewidth= 3, label= 'media')
plt.axvline(titanic.Age.median(), c= 'green', linewidth = 3, label= 'mediana')
plt.axvline(24, c= 'orange', linewidth = 3, label= 'moda')
plt.title('Summary statistics')
plt.legend()
```

#### Plot the correlation matrix using seaborn.

```python
sns.set(style='white')    # estilo blanco

mascara=np.triu(np.ones_like(titanic.corr(), dtype=bool))   # genera una mascara

cmap=sns.diverging_palette(0, 10, as_cmap=True)   # paleta de colores


sns.heatmap(titanic.corr(),
            mask=mascara,
            cmap=cmap,
            vmax=1,
            vmin=-1,
            center=0,
            square=True,
            linewidth=0.5,
            cbar_kws={'shrink': 0.5},
            annot=True
           );
```

#### What are the most correlated features?

```python
"""
Pclass with Fare, Age with Pclass
"""
```
